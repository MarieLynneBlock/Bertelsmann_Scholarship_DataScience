#### Data 3

Risk Model dataset:
This dataset includes 2,392,998 screening mammograms (called the "index mammogram") from women included in the Breast Cancer Surveillance Consortium. All women did not have a previous diagnosis of breast cancer and did not have any breast imaging in the nine months preceding the index screening mammogram. However, all women had undergone previous breast mammography in the prior five years (though not in the last nine months). Cancer registry and pathology data were linked to the mammography data and incident breast cancer (invasive or ductal carcinoma in situ) within one year following the index screening mammogram was assessed.

To reduce the size of the dataset, the data have been aggregated by the cross-classification of risk factors and outcome with a count indicating the frequency of each combination. This reduces the dataset to 280,660 records. The variables in the ASCII dataset "risk.txt" are described in the table.

Data from: http://www.bcsc-research.org/rfdataset/dataset.html

Why use this data? My case study is going to be on Singapore women compared to the average as collected by BCSC.

Filtering criteria for original data to be similar to Sg data:
<br> 1. Age group between 50-69 (agegrp = 4,5,6,7)
<br> 2. race is Asian (race=2)

Potential limitations:
<br> 1. BCSC data recorded in 1996-2002, not as updated till 2015 in Singapore's data
<br> 2. Sg dataset incomplete, does not have data such as family history, bmi etc
<br> 3. BCSC data is for all women who had undergone previous breast mammography in the prior five years (though not in the last nine months), cannot compare those with and without mammogram to cancer detection

Though of course mammogram is able to detect breast cancer up to 2 years before it can be felt (https://www.cancercare.org/publications/82-early_detection_and_breast_cancer)

```python
bcsc_filter = bcsc[(bcsc['race']==2) & (bcsc['agegrp'].isin(range(4,8)))]
bcsc_filter.head()
```

```python
bcsc_filter.shape
```

```python
bcsc_filter.describe()
```

```python
bcsc_filter['count'].sum()
```

```python
#percentage of people with breast cancer (invasive and ductal carcinoma in situ breast cancer)
bcsc_filter.groupby('cancer').agg({'count':'sum'})/bcsc_filter['count'].sum()
```

```python
#percentage with invasive breast cancer
bcsc_filter.groupby('invasive').agg({'count':'sum'})/bcsc_filter['count'].sum()
```

using the invasive cancer column because some DCIS might not require treatment. a problem known as overdiagnosis and over treatment https://www.cancer.gov/types/breast/mammograms-fact-sheet#q3

#### Mammogram and cancer detection
Lesser percentage of people detected with breast cancer during the index screening had negative results during their last mammogram, higher rate of false positives amongst them as well

```python
#result of last mammogram has slight impact on cancer detection during index screening
bcsc_filter.groupby(['invasive','lastmamm']).agg({'count':'sum'})/bcsc_filter.groupby(['invasive']).agg({'count':'sum'})
```

```python
bcsc_filter.groupby(['invasive','lastmamm']).agg({'count':'sum'})
```

potential problem: high rate of false positives,
More than 50% of women screened annually for 10 years in the United States will experience a false-positive result, and many of these women will have a biopsy.
https://www.cancer.gov/types/breast/mammograms-fact-sheet#q3

#### Other factors that may affect breast cancer detection

```python
#menopause
bcsc_filter.groupby(['invasive','menopause']).agg({'count':'sum'})/bcsc_filter.groupby(['invasive']).agg({'count':'sum'})
```

```python
#bmi
bcsc_filter.groupby(['invasive','bmi']).agg({'count':'sum'})/bcsc_filter.groupby(['invasive']).agg({'count':'sum'})
```

```python
#previous breast procedure
bcsc_filter.groupby('invasive')['brstproc'].value_counts()/bcsc_filter.groupby('invasive')['brstproc'].count()
```

```python
#number of first degree relatives with breast cancer
bcsc_filter.groupby(['invasive','nrelbc']).agg({'count':'sum'})/bcsc_filter.groupby(['invasive']).agg({'count':'sum'})
# bcsc_filter.groupby('invasive')['nrelbc'].value_counts()/bcsc_filter.groupby('invasive')['nrelbc'].count()
```

```python
(bcsc_filter.groupby(['invasive','nrelbc']).agg({'count':'sum'})/bcsc_filter.groupby(['invasive']).agg({'count':'sum'})).plot(kind='bar')
```
